# 2. Simple Linear Regression

##

```python
from sklearn.linear_model import LinearRegression
```

```python
dfs = [df_distribution, df_attempts, df_goals]

df_merged = reduce(lambda left, right: pd.merge(left, right, on='id_player', how='left'), dfs)
df_merged.dropna().head(5)
```

```python
import statsmodels.api as sm
```

```python
y = df_merged["goals"].dropna()
X = df_merged[["passes_attempted", "matches_appearance"]].dropna()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
X = sm.add_constant(X)  # This is required for statsmodels
model = sm.OLS(y, X).fit()
model.summary()
```
